## Handle class imbalance

### Subtask:
Address the extreme class imbalance in the dataset using appropriate techniques such as oversampling, undersampling, or a combination.


**Reasoning**:
Check the distribution of the target variable 'Class' to quantify the class imbalance using value_counts and visualize it using a bar plot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check the distribution of the target variable 'Class'
class_distribution = df['Class'].value_counts()
print("Class distribution:\n", class_distribution)

# Visualize the class distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=class_distribution.index, y=class_distribution.values)
plt.title('Distribution of Target Variable (Class)')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks(ticks=class_distribution.index, labels=['Normal (0)', 'Prediabetes (1)', 'Diabetes (2)'])
plt.show()
```

**Reasoning**:
Apply the SMOTE oversampling technique to handle the class imbalance since it is effective for multiclass problems and the dataset size is relatively small. Then, verify the class distribution after applying SMOTE.



```python
from imblearn.over_sampling import SMOTE

# Separate features and target variable
X = df.drop('Class', axis=1)
y = df['Class']

# Apply SMOTE to the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Verify the class distribution after applying SMOTE
resampled_class_distribution = y_resampled.value_counts()
print("Class distribution after SMOTE:\n", resampled_class_distribution)

# Visualize the resampled class distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=resampled_class_distribution.index, y=resampled_class_distribution.values)
plt.title('Distribution of Target Variable (Class) After SMOTE')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks(ticks=resampled_class_distribution.index, labels=['Normal (0)', 'Prediabetes (1)', 'Diabetes (2)'])
plt.show()
```

## Data preprocessing

### Subtask:
Prepare the data for machine learning models. This may involve handling missing values, encoding categorical features, and scaling numerical features.


**Reasoning**:
Identify numerical columns, scale them using StandardScaler, and prepare the data for model training.



```python
from sklearn.preprocessing import StandardScaler

# Identify numerical columns (all columns except 'Class' in X_resampled are numerical)
numerical_cols = X_resampled.columns

# Initialize StandardScaler
scaler = StandardScaler()

# Scale the numerical features
X_scaled = scaler.fit_transform(X_resampled[numerical_cols])

# Convert the scaled array back to a DataFrame for easier handling
X_scaled_df = pd.DataFrame(X_scaled, columns=numerical_cols)

# Display the first few rows of the scaled DataFrame
print("Scaled features (first 5 rows):")
display(X_scaled_df.head())
```

## Feature selection

### Subtask:
Select relevant features that contribute most to predicting diabetes status.


**Reasoning**:
Import the necessary feature selection module and scoring function.



```python
from sklearn.feature_selection import SelectKBest, f_classif
```

**Reasoning**:
Instantiate SelectKBest, fit it to the scaled data and resampled target, get the selected feature names, and create a new DataFrame with the selected features.



```python
# Instantiate SelectKBest to select half of the features
k = X_scaled_df.shape[1] // 2
selector = SelectKBest(score_func=f_classif, k=k)

# Fit the selector to the scaled data and resampled target
selector.fit(X_scaled_df, y_resampled)

# Get the indices of the selected features
selected_feature_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_feature_names = X_scaled_df.columns[selected_feature_indices]

# Create a new DataFrame with only the selected features
X_selected = X_scaled_df[selected_feature_names]

# Display the first few rows of the new DataFrame
print("Selected features (first 5 rows):")
display(X_selected.head())

# Print the names of the selected features
print("\nNames of selected features:")
print(selected_feature_names)
```

## Model selection and training

### Subtask:
Choose appropriate machine learning models for multiclass classification and train them on the preprocessed data.


**Reasoning**:
Import the necessary classification models and the train-test split function.



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

**Reasoning**:
Split the data into training and testing sets, initialize the models, and train them.



```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_resampled, test_size=0.2, random_state=42)

# Initialize the models
log_reg_model = LogisticRegression(random_state=42, max_iter=1000)
rf_model = RandomForestClassifier(random_state=42)
svm_model = SVC(random_state=42)

# Train the models
log_reg_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)

print("Models trained successfully.")
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained models using relevant metrics such as accuracy, precision, recall, F1-score, and confusion matrix.


**Reasoning**:
Evaluate the performance of the trained models using relevant metrics.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Evaluate Logistic Regression model
log_reg_pred = log_reg_model.predict(X_test)
log_reg_accuracy = accuracy_score(y_test, log_reg_pred)
log_reg_precision = precision_score(y_test, log_reg_pred, average='weighted')
log_reg_recall = recall_score(y_test, log_reg_pred, average='weighted')
log_reg_f1 = f1_score(y_test, log_reg_pred, average='weighted')
log_reg_conf_matrix = confusion_matrix(y_test, log_reg_pred)

print("Logistic Regression Model Evaluation:")
print(f"Accuracy: {log_reg_accuracy:.4f}")
print(f"Precision (weighted): {log_reg_precision:.4f}")
print(f"Recall (weighted): {log_reg_recall:.4f}")
print(f"F1-score (weighted): {log_reg_f1:.4f}")
print("Confusion Matrix:")
print(log_reg_conf_matrix)
print("-" * 30)

# Evaluate Random Forest model
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred, average='weighted')
rf_recall = recall_score(y_test, rf_pred, average='weighted')
rf_f1 = f1_score(y_test, rf_pred, average='weighted')
rf_conf_matrix = confusion_matrix(y_test, rf_pred)

print("Random Forest Model Evaluation:")
print(f"Accuracy: {rf_accuracy:.4f}")
print(f"Precision (weighted): {rf_precision:.4f}")
print(f"Recall (weighted): {rf_recall:.4f}")
print(f"F1-score (weighted): {rf_f1:.4f}")
print("Confusion Matrix:")
print(rf_conf_matrix)
print("-" * 30)

# Evaluate SVM model
svm_pred = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_pred)
svm_precision = precision_score(y_test, svm_pred, average='weighted')
svm_recall = recall_score(y_test, svm_pred, average='weighted')
svm_f1 = f1_score(y_test, svm_pred, average='weighted')
svm_conf_matrix = confusion_matrix(y_test, svm_pred)

print("SVM Model Evaluation:")
print(f"Accuracy: {svm_accuracy:.4f}")
print(f"Precision (weighted): {svm_precision:.4f}")
print(f"Recall (weighted): {svm_recall:.4f}")
print(f"F1-score (weighted): {svm_f1:.4f}")
print("Confusion Matrix:")
print(svm_conf_matrix)
print("-" * 30)
```

## Hyperparameter tuning

### Subtask:
Optimize the performance of the best-performing model by tuning its hyperparameters.


**Reasoning**:
Identify the best performing model from the evaluation results and import GridSearchCV to prepare for hyperparameter tuning.



```python
from sklearn.model_selection import GridSearchCV

# Based on the previous evaluation, Random Forest was the best performing model.
best_model = rf_model

print("Identified Random Forest as the best performing model.")
```

**Reasoning**:
Define the hyperparameter grid for Random Forest and instantiate GridSearchCV with the model, grid, and cross-validation settings.



```python
# Define the hyperparameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=best_model, param_grid=param_grid, cv=5, scoring='f1_weighted', n_jobs=-1)

print("Hyperparameter grid defined and GridSearchCV instantiated.")
```

**Reasoning**:
Fit GridSearchCV to the training data to find the best hyperparameters and then print the best parameters found.



```python
# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

print("Grid search completed. Best hyperparameters found:")
print(best_params)
```

**Reasoning**:
Train a new Random Forest model with the best hyperparameters found by GridSearchCV and evaluate its performance on the test set.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Train a new Random Forest model with the best hyperparameters
optimized_rf_model = RandomForestClassifier(**best_params, random_state=42)
optimized_rf_model.fit(X_train, y_train)

# Evaluate the optimized model on the test set
optimized_rf_pred = optimized_rf_model.predict(X_test)
optimized_rf_accuracy = accuracy_score(y_test, optimized_rf_pred)
optimized_rf_precision = precision_score(y_test, optimized_rf_pred, average='weighted')
optimized_rf_recall = recall_score(y_test, optimized_rf_pred, average='weighted')
optimized_rf_f1 = f1_score(y_test, optimized_rf_pred, average='weighted')
optimized_rf_conf_matrix = confusion_matrix(y_test, optimized_rf_pred)

print("\nOptimized Random Forest Model Evaluation:")
print(f"Accuracy: {optimized_rf_accuracy:.4f}")
print(f"Precision (weighted): {optimized_rf_precision:.4f}")
print(f"Recall (weighted): {optimized_rf_recall:.4f}")
print(f"F1-score (weighted): {optimized_rf_f1:.4f}")
print("Confusion Matrix:")
print(optimized_rf_conf_matrix)
```

## Mathematical Interpretation of Model Evaluation

Based on the evaluation metrics and the confusion matrix, here's a mathematical interpretation of the optimized Random Forest model's performance:

*   **Accuracy:** This is the proportion of correctly classified instances out of the total number of instances. Mathematically, it's calculated as:
    $$ \text{Accuracy} = \frac{\text{True Positives} + \text{True Negatives}}{\text{True Positives} + \text{True Negatives} + \text{False Positives} + \text{False Negatives}} $$
    In our case, the optimized Random Forest model has an accuracy of {optimized_rf_accuracy:.4f}, meaning that approximately {optimized_rf_accuracy:.2%} of the predictions made by the model on the test set were correct.

*   **Precision:** This metric tells us how many of the instances predicted as positive for a certain class actually belong to that class. For a specific class (let's say Class C), it's calculated as:
    $$ \text{Precision}_C = \frac{\text{True Positives}_C}{\text{True Positives}_C + \text{False Positives}_C} $$
    The weighted average precision of {optimized_rf_precision:.4f} indicates the model's ability to avoid falsely classifying a negative instance as positive, averaged across all classes and weighted by the number of instances in each class.

*   **Recall (Sensitivity):** This metric measures the proportion of actual positive instances of a class that were correctly identified by the model. For a specific class (Class C), it's calculated as:
    $$ \text{Recall}_C = \frac{\text{True Positives}_C}{\text{True Positives}_C + \text{False Negatives}_C} $$
    The weighted average recall of {optimized_rf_recall:.4f} indicates the model's ability to find all the positive instances of each class, averaged and weighted.

*   **F1-score:** This is the harmonic mean of precision and recall, providing a single metric that balances both. For a specific class (Class C), it's calculated as:
    $$ \text{F1-score}_C = 2 \times \frac{\text{Precision}_C \times \text{Recall}_C}{\text{Precision}_C + \text{Recall}_C} $$
    The weighted average F1-score of {optimized_rf_f1:.4f} gives a combined measure of the model's precision and recall, weighted by class support.

*   **Confusion Matrix:** As we visualized, the confusion matrix provides a detailed breakdown of predictions. Each cell $(i, j)$ in the matrix shows the number of instances that are actually in class $i$ but were predicted as class $j$.
    *   The diagonal elements represent the **True Positives** for each class (correct predictions).
    *   The off-diagonal elements represent the **False Positives** (column-wise, instances incorrectly predicted as a certain class) and **False Negatives** (row-wise, instances of a class incorrectly predicted as another).
